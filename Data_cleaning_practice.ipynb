```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv("data_cleaning_challenge.csv")
df
```

### Required file output
<img src='./images/final_goal.png'>

```python
# Drop the extra columns
df = df.drop(columns=["Unnamed: 9", "Unnamed: 10"])
df
```

```python
# Drop nan in column 'Row Type'
df = df[df["Row Type"].notna()]
df.head()
```

```python
# Count whole users amount -> calculate how many user based on 'Row Type'
count_values = []
amount = 0
for ele in df["Row Type"]:
    if 'first name' in ele:
       amount += 1
    count_values.append(amount)
```

```python
# Add new column on df
df['Iteration'] = count_values
df.head(11)
```

## Split into two dataset

### Dataframe1 with name

```python
df_name = df.loc[df["Row Type"].str.contains("first name")]
```

```python
cols = list(df_name.columns)
df_name = df_name[cols[:3] + [cols[-1]]]
df_name
```

```python
# Rename columns' name
renamed_cols = ["First Name", "Last Name", "Date", "Iteration"]
df_name.columns = renamed_cols
```

```python
# Extract data we need
df_name[renamed_cols[:3]] = df_name[renamed_cols[:3]].applymap(lambda x: x.split(":")[1])
```

```python
df_name.head(10)
```

### Dataframe2 with other information

```python
df_info = df.loc[~(df["Row Type"].str.contains("first name"))]
df_info
```

```python
# Drop the unnecessary info
df_info = df_info.loc[~(df_info["Row Type"].str.contains(r"Row Type|Total"))]
df_info.head()
```

## Join two data sets on Iteration

```python
df_output = pd.merge(df_name, df_info, on='Iteration')
df_output
```

```python
df_output.to_csv("data_cleaning_output.csv", index=False)
```
